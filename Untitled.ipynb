```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
# 1 count of male and female
```

```python
data['gender'].value_counts()
```

```python
# 2 based on parental level of education
```

```python
data['parental level of education'].value_counts()
```

```python
values=data.groupby('parental level of education').mean()
print(values)
```

```python
# 3 average score based on math,reading and writing based on gender and test preparation course
```

```python
values=data.groupby('gender')[['math score','reading score','writing score']].mean()
print(values)
```

```python
values=data.groupby('test preparation course')[['math score','reading score','writing score']].mean()
print(values)
```

```python
# 4 scoring variation for math,reading and writing based on gender and test preparation course
```

```python
values=data.groupby('gender').var()
print(values)
```

```python
values=data.groupby('gender')[['math score','reading score','writing score']].var()
print(values)
```

```python
values=data.groupby('test preparation course')[['math score','reading score','writing score']].var()
print(values)
```

```python
#5 sorting top 25 students based on math score
```

```python
values=data.sort_values(by=['math score','reading score','writing score'],ascending=False)
print(values)
```
